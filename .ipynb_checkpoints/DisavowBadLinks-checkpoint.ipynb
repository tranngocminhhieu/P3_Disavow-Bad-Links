## Converting URL in sitemap list to simple path

```python
# Converting URL in sitemap list to simple path
from urllib.parse import urlparse # For Extracting path in URL
for row in list_sitemap:
    row[0] = urlparse(row[0]).path.replace('en/','').replace('/','')
print(pd.DataFrame(list_sitemap))
```

## Converting the datasets to list of lists

```python
dataset_ahref = explore_dataset_ahref.values.tolist()
dataset_console = explore_dataset_console.values.tolist()
dataset_moz = explore_dataset_moz[7:]

print('First data row in Ahref:')
print(dataset_ahref[0])
print('\nFirst data row in Console:')
print(dataset_console[0])
print('\nFirst data row in Moz:')
print(dataset_moz[0])
```

## Extracting the required data for datasets

```python
# We need install tldextract module before import it
import tldextract # For Extracting TLD domain (Root domain)

# Extracting the required data for Ahref
dataset_ahref_extracted = []

for row in dataset_ahref:
    page = row[1]
    domain = tldextract.extract(page).registered_domain
    dr = float(row[5])
    path = urlparse(page).path
    url = urlparse(page).netloc + urlparse(page).path
    data = [domain,url,path,dr]
    if data not in dataset_ahref_extracted and domain != '':
        dataset_ahref_extracted.append(data)

# Extracting the required data for Moz
dataset_moz_extracted = []

for row in dataset_moz:
    page = 'https://'+row[0]
    domain = tldextract.extract(page).registered_domain
    da = float(row[4])
    ss = row[3].replace('%','')
    if ss == '--':
        ss = 0.0
    else:
        ss = float(ss)
    path = urlparse(page).path
    url = urlparse(page).netloc + urlparse(page).path
    data = [domain,url,path,da,ss]
    if data not in dataset_moz_extracted and domain != '':
        dataset_moz_extracted.append(data)

# Extracting the domains for Console
dataset_console_domain = []

for row in dataset_console:
    domain = row[0]
    if domain not in dataset_console_domain:
        dataset_console_domain.append([domain])

pd.DataFrame(dataset_ahref_extracted)
```

## Grouping all datesets

### Step 1: Grouping distinct domains, URL, Path

```python
dataset_ahref_moz_domain = []

for row in dataset_ahref_extracted + dataset_moz_extracted:
    domain = row[0]
    if domain not in dataset_ahref_moz_domain:
        dataset_ahref_moz_domain.append(domain)

print(pd.DataFrame(dataset_ahref_moz_domain))
```

```python
dataset_group_domain_url_path = []

for row in dataset_ahref_extracted + dataset_moz_extracted:
    domain = row[0]
    url = row[1]
    path = row[2]
    data = [domain,url,path]
    if data not in dataset_group_domain_url_path:
        dataset_group_domain_url_path.append(data)

for domain in dataset_console_domain:
    if domain not in dataset_ahref_moz_domain:
        dataset_group_domain_url_path.append([domain[0],'None','None'])

dataset_group_domain_url_path = sorted(dataset_group_domain_url_path)

pd.DataFrame(dataset_group_domain_url_path)
```

### Step 2: Grouping all distinct domain with it's data
Domain | URL | Path | DR | DA | SS | Disavow

```python
dataset_group = []
for row in dataset_group_domain_url_path:
    domain = row[0]
    url = row[1]
    path = row[2]
    data = [domain,url,path,'None','None','None','domain:'+domain]
    for a_row in dataset_ahref_extracted:
        a_domain = a_row[0]
        a_dr = a_row[3]
        if domain == a_domain:
            data[3] = a_dr
    for m_row in dataset_moz_extracted:
        m_domain = m_row[0]
        m_da = m_row[3]
        m_ss = m_row[4]
        if domain == m_domain:
            data[4] = m_da
            data[5] = m_ss
    dataset_group.append(data)

pd.DataFrame(dataset_group)
```

## Extracting bad domains to Disavow list and DMCA report list

**Our conditions for Disavow:**
- DR < 30
- DA < 10
- DA < 40 and SS > DA*20% (Exp: DA = 25, SS > 5)
- DA >= 40 and SS > DA*40% (Exp: DA = 50, SS > 20)
- Domain or path contain bad keywords
- Path match with path in sitemap

**Our conditions for DMCA report:**
- Path match with path in sitemap

```python
## Checking bad keywords in domain
def is_bad_domain(domain):
    for row in list_key_domain:
        keyword = row[0]
        if '.' in keyword: # Checking suffix exactly ('.in' keyword will not match '.info' suffix)
            if tldextract.extract(domain).suffix in keyword:
                return True
        elif keyword in domain:
            return True
    return False

## Checking bad keywords in path
def is_bad_path(path):
    for row in list_key_path:
        keyword = row[0]
        if keyword in path:
            return True
    return False

## Checking copied path
def is_copied_path(path):
    for row in list_sitemap:
        post = row[0]
        if post in path:
            return True
    return False

print(is_bad_domain('abc.pw'))
print(is_bad_path('domain-list.html'))
print(is_copied_path('/digital-marketing/chuyen-doi-tawk-to.html'))
```

```python
# Extracting bad domains
disavow = []
manual_check = []
dmca_report = []

## Extracting bad domain/path with conditions
for row in dataset_group:
    domain = row[0]
    path = row[2]
    dr = row[3]
    da = row[4]
    ss = row[5]
    dis = [row[6],dr,da,ss]
    if [domain] in list_whitelist:
        pass
    elif is_copied_path(path):
        disavow.append(dis)
        dmca_report.append(row)
    elif is_bad_domain(domain):
        disavow.append(dis)
    elif is_bad_path(path):
        disavow.append(dis)
    elif dr != 'None' and dr < 30:
        disavow.append(dis)
    elif da != 'None' and da < 10:
        disavow.append(dis)
    elif ss != 'None' and da !='None' and da < 40 and ss > da * 0.2:
        disavow.append(dis)
    elif ss != 'None' and da !='None' and da >= 40 and ss > da * 0.4:
        disavow.append(dis)
    else:
        manual_check.append(row)

disavow_distinct = pd.DataFrame(disavow).drop_duplicates().values.tolist()
```
